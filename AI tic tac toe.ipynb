```python
'''creating a tic tac toe game invloves some steps they are
1.printing the game board
2.take player input
3.check for win or tie
4.switch the player
5.check the win or tie game'''
import random

# Function to draw the Tic-Tac-Toe board
def draw_board(board):
    print(" " + board[0] + " | " + board[1] + " | " + board[2])
    print("---+---+---")
    print(" " + board[3] + " | " + board[4] + " | " + board[5])
    print("---+---+---")
    print(" " + board[6] + " | " + board[7] + " | " + board[8])

# Function to check for a win
def check_win(board, mark):
    win_conditions = [
        [0, 1, 2], [3, 4, 5], [6, 7, 8],  # Rows
        [0, 3, 6], [1, 4, 7], [2, 5, 8],  # Columns
        [0, 4, 8], [2, 4, 6]  # Diagonals
    ]
    for condition in win_conditions:
        if board[condition[0]] == board[condition[1]] == board[condition[2]] == mark:
            return True
    return False

# Function to make the computer's move
def computer_move(board, computer_mark, player_mark):
    # Check if the computer can win in the next move
    for i in range(9):
        if board[i] == ' ':
            board[i] = computer_mark
            if check_win(board, computer_mark):
                return i
            else:
                board[i] = ' '

    # Check if the player can win in the next move and block them
    for i in range(9):
        if board[i] == ' ':
            board[i] = player_mark
            if check_win(board, player_mark):
                return i
            else:
                board[i] = ' '

    # Choose a random move
    empty_cells = [i for i in range(9) if board[i] == ' ']
    if empty_cells:
        return random.choice(empty_cells)

    return None

# Function to play the game
def play_game():
    board = [' '] * 9
    player_mark = 'X'
    computer_mark = 'O'

    print("Welcome to Tic-Tac-Toe!")
    print("You are playing as X. Let's begin!")

    while True:
        draw_board(board)
        print("It's your turn. Enter the position (1-9): ")
        player_move = int(input()) - 1

        if board[player_move] != ' ':
            print("Invalid move. Try again.")
            continue

        board[player_move] = player_mark

        if check_win(board, player_mark):
            draw_board(board)
            print("Congratulations! You won!")
            break

        if ' ' not in board:
            draw_board(board)
            print("It's a draw!")
            break

        print("It's the computer's turn.")
        computer_move_index = computer_move(board, computer_mark, player_mark)
        board[computer_move_index] = computer_mark

        if check_win(board, computer_mark):
            draw_board(board)
            print("Sorry! You lost.")
            break

        if ' ' not in board:
            draw_board(board)
            print("It's a draw!")
            break

# Start the game
play_game()
```
